```python
import pandas as pd

def read_usuarios_csv(filename='usuarios.csv'):

   # Lê os dados de um arquivo CSV e retorna um DataFrame.
   #    Args:
   #      filename (str): O nome do arquivo CSV a ser lido.
   #
   # Returns:
   #     pd.DataFrame: Um DataFrame contendo os dados do arquivo CSV.
   #                   Retorna Estrutura Padrao se o arquivo não for encontrado.


    try:
        df = pd.read_csv(filename)
        return df
    except FileNotFoundError:
        print(f"Erro: O arquivo '{filename}' não foi encontrado. Utilizando estrutura interna")

        ids = [10000 + i for i in range(10)]
        df = pd.DataFrame({'ID': ids})

        return df

    except Exception as e:
        print(f"Ocorreu um erro ao ler o arquivo: {e}")
        return None

df_usuarios = read_usuarios_csv('usuariosX.csv')

print(df_usuarios)

#if df_usuarios is not None:
#     print(df_usuarios.head())
```

```python
def adicionar_nome_e_incrementar(df):

    #Adiciona uma nova coluna 'nome' a cada registro do DataFrame
    #e retorna um novo DataFrame com a informação incrementada.

    df_incrementado = df.copy() # Cria uma cópia para não modificar o original

    nomes_gerados = [f"Usuario_{i+1}" for i in range(len(df_incrementado))]
    df_incrementado['nome'] = nomes_gerados

    return df_incrementado


df_usuarios_incrementada = adicionar_nome_e_incrementar(df_usuarios)
print("DataFrame original df_usuarios:")
print("\nNovo DataFrame df_usuarios_incrementada:")
print(df_usuarios_incrementada)

```

```python
import pandas as pd

# Salvar o DataFrame df_usuarios_incrementada em um novo arquivo CSV
output_filename = 'usuarios_incrementados.csv'
df_usuarios_incrementada.to_csv(output_filename, index=False)

print(f"DataFrame salvo como '{output_filename}'")
```
